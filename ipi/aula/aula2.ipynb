# Negativo de imagem acromática

Fórmula: g(x,y) = 255 - f(x,y) por broadcast

```python
neg = 255 - img # numpy
print(neg.shape)
```

```python
plt.imshow(neg, cmap='gray')
```

```python
fig,axs = plt.subplots(1, 2)
axs[0].set_title('Original')
axs[0].axis('off')
axs[0].imshow(img, cmap='gray')
axs[1].set_title('Negativo')
axs[1].imshow(neg, cmap='gray')
axs[1].axis('off')
plt.show()
```

## Ajuste de Brilho em imagens Acromáticas

```python
img = plt.imread('breast_digital_Xray.tif') # retorna um array numpy
img.shape
```

```python
c = 100
img_mais_brilho = img.astype(np.float64) + c
# trunca valores saturados
img_mais_brilho[img_mais_brilho > 255] = 255
img_mais_brilho = img_mais_brilho.astype(np.uint)
```

```python
fig,axs = plt.subplots(1, 2)
axs[0].set_title('Original')
axs[0].axis('off')
axs[0].imshow(img, cmap='gray')
axs[1].set_title('Mais Brilho')
axs[1].imshow(img_mais_brilho, cmap='gray')
axs[1].axis('off')
plt.show()
```

```python
c = 100
img_menos_brilho = img.astype(np.float64) - c
# trunca valores saturados
img_menos_brilho[img_menos_brilho < 0] = 0
img_menos_brilho = img_menos_brilho.astype(np.uint)
```

```python
fig,axs = plt.subplots(1, 2)
axs[0].set_title('Original')
axs[0].axis('off')
axs[0].imshow(img, cmap='gray')
axs[1].set_title('Menos Brilho')
axs[1].imshow(img_menos_brilho, cmap='gray')
axs[1].axis('off')
plt.show()
```
