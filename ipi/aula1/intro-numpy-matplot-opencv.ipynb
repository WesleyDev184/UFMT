# Introdução ao `numpy`, matplotlib e opencv

A biblioteca `numpy` é base para construção de outras bibliotecas, tais como matplotlib. Para trabalhar com `numpy`, primeiro precisamos importa-la. 

```python
import numpy as np 
```

Na linha anterior, nomeamos o pacote `numpy` como `np`, uma abreviação mais fácil de usar. 

Na biblioteca `numpy`, um *array* é um termo genérico para definir um tensor multidimensional. Por exemplo, a seguir, usamos a função `np.array()` para criar dois arrays unidimensionais, ou seja, dois vetores, os quais serão referenciados por `x` e `y`.

```python
x = np.array([3, 4, 5])
y = np.array([4, 9, 7])
```

Note que `np.array()` recebe uma lista `Python` como argumento. Note também que se você esquecer de rodar o comando `import numpy as np`, então a chamada à função `np.array()` resultará em um erro. A sintax `np.array()` indica que a função é parte do pacote `numpy`, abreviado como `np`. 

Uma vez que as variáveis `x` e `y` agora referenciam arrays `numpy`, quando realizamos a soma usando o operador `+` obtemos um resultado bastante intuitivo, a saber: os arrays são somados componente-por-componente. Isso é possível pois os arrays `numpy` são tipados e homogêneos. 

```python
z = x + y
z
```

Em `numpy`, matrizes são tipicamente representadas por arrays bidimensionais. O exemplo a seguir mostra como criar uma matriz (array 2D) usando numpy. 

```python
x = np.array([[1, 2], [3, 4]])
x
```

Note que, nesse caso, `np.array()` recebe uma lista bidimensional como argumento. O objeto numpy referenciado por `x` tem vários 
*atributos*. Para acessá-los, digitamos `x.attribute`, substituindo a palavra `attribute` pelo nome do atributo desejado. 
Por exemplo, o código a seguir acessa o atributo `ndim` (número de dimensões do array) de  `x`. 

```python
x.ndim
```

A saída indica que `x` referencia um array bidimensional. Similarmente, `x.dtype` indica o tipo de dado (*data type*) armazenado no array referenciado por `x`. Veja a seguir que `x` referencia um array composto de inteiros de 64 bits:

```python
x.dtype
```

Mas, por que o array referenciado por `x` possui valores inteiros? Isso acontece porque criamos `x` passando apenas inteiros para a função `np.array()`. Se passarmos valores decimais, obtemos um array de *números de pontos flutuante* (aproximações para os números reais). 

```python
np.array([[1, 2], [3.0, 4]]).dtype
```

Podemos especificar o tipo do dado passando um argumento `dtype` para `np.array()`.

```python
y = np.array([[1, 2], [3, 4]], dtype=np.float64)
```

Podemos descobrir o números de linhas e colunas olhando o atributo `shape`, que devolve uma tupla com as dimensões.

```python
y.shape
```

Um *método* é uma função encapsulada por um objeto `Python`. Por exemplo, dado um objeto array referenciado por `x`, a expressão `x.sum()` invoca o método `sum()` que, por sua vez, soma todos os elementos do array. 

```python
x = np.array([1, 2, 3, 4])
x.sum()
```

Podemos também somar os elementos do array referenciado por `x` passando `x` como argumento da função `np.sum()`. 

```python
x = np.array([1, 2, 3, 4])
np.sum(x)
```

Como outro exemplo, o método `reshape()` retorna outro array com os mesmos elementos do array referenciado por `x`, mas com um formato diferente. Para explicitar o formato desejado, passamos como argumento para `reshape()` uma `tupla` indicando o novo formato. Por exemplo, se passarmos a tupla `(2, 3)`, especificamos que gostaríamos de criar um array bidimensional com $2$ linhas e $3$ colunas. No exemplo a seguir um array 1D é "convertido" para um array 2D.

```python
x = np.array([1, 2, 3, 4, 5, 6])
print('Inicialmente x:\n', x)
x_reshaped = x.reshape((2, 3))
print('Depois da modificação x:\n', x_reshaped)
```

A saída anterior revela que matrizes numpy são especificadas como uma sequência de linhas. 
Isso é chamado de ordenação por linha, em oposição à ordenação por coluna. Em `Python` (e `numpy`) a indexação começa em 0. Isso significa que o elemento superior esquerdo de `x_reshaped` é acessado com o índice `[0,0]`.

```python
x_reshaped[0, 0] 
```

Similarmente, `x_reshaped[1,2]` retorna o valor do elemento na segunda linha e terceira coluna de `x_reshaped`. 

```python
x_reshaped[1, 2] 
```

Similarmente, `x[2]` devolve a terceira entrada de `x`. 

```python
x[2]
```

Agora, observe o que acontece quando modificamos o elemento superior esquerdo de `x_reshaped`.  Para nossa surpresa, descobrimos que o primeiro elemento de `x` também é modificado. Isso acontece porque os dois objetos ocupam as mesmas posições da memória.

```python
print('x antes de modificar x_reshaped:\n', x)
print('x_reshaped antes de modificar x_reshaped:\n', x_reshaped)
x_reshaped[0, 0] = 5
print('x_reshaped depois de modificar o elemento superior esquerdo:\n', x_reshaped)
print('x depois de modificar o elemento superior esquerdo de x_reshaped:\n', x)
```

Uma operação importante envolvendo arrays é a transposição. O atributo `T` de um array devolve o array transposto. 

```python
x_reshaped.T
```

Para computar a raiz quadrada de cada elemento de um array, usamos a função `np.sqrt()`: 

```python
np.sqrt(x)
```

Podemos também calcular o quadrado de cada elemento de um array com o operador `**`.

```python
x ** 2
```

### Geração de números randômicos

Em muitas situações, precisamos gerar arrays com números amostrados de alguma distribuição. Por exemplo,
a função `np.random.normal()` gera um array com valores amostrados de uma distribuição normal. 
Podemos aprender muito sobre essa função olhando seu manual de ajuda via `np.random.normal?`. 

A primeira linha da página de ajuda mostra a assinatura da função `np.normal(loc=0.0, scale=1.0, size=None)`. A assinatura nos diz que a função tem argumentos definidos com *keyword* (palavra-chave), o que significa que quando eles são passados para a função, eles podem ser referenciados pelo nome (em qualquer ordem). `Python` usa também argumentos posicionais, os quais não precisam de palavra-chave. Por default, esta função amostra valores de uma distribuição normal com média `loc` (default $0$) e desvio padrão `scale` (default $1$). Além disso, por default, um único valor é amostrado, a menos que o argumento `size` seja alterado. No exemplo a seguir amostraremos $50$ números de uma distribuição normal $N(0,1)$. 

```python
x = np.random.normal(size=50)
x
```

Para obter valores independentes de uma distribuição normal com média $25$ e desvio padrão $1$  (y ~ $N(25,1)$):

```python
y = np.random.normal(loc=25, scale=1, size=50)
y
```

A função `np.corrcoef()` computa a matriz de correlação entre `x` and `y`.  

```python
np.corrcoef(x, y)
```

Você provavelmente notou que se reexecutar os comandos anteriores obterá um conjunto diferente de resultados. Em particular, cada vez que chamarmos `np.random.normal()`, obteremos uma resposta diferente, conforme mostrado no exemplo a seguir.

```python
print(np.random.normal(scale=5, size=2))
print(np.random.normal(scale=5, size=2)) 
```

Para garantir que o código fornecerá exatamente os mesmos resultados cada vez que for executado, podemos definir uma semente usando a função `np.random.default_rng()`. Esta função recebe um argumento inteiro arbitrário especificado pelo usuário e retorna um objeto gerador de números randômicos inicializado com a semente. Se definirmos a mesma semente antes de gerar dados aleatórios, a nova execução do código produzirá os mesmos resultados. O objeto retornado possui essencialmente todos os métodos de geração de números aleatórios encontrados em `np.random`. Portanto, para gerar valores amostrados de uma distribuição normal usamos `rng.normal()`.

```python
rng = np.random.default_rng(1303)
print(rng.normal(scale=5, size=2))
rng2 = np.random.default_rng(1303)
print(rng2.normal(scale=5, size=2)) 
```

As funções `np.mean()`,  `np.var()`, and `np.std()` podem ser usadas para calcular a média, variância, e desvio padrão de arrays. Essas funções estão também disponíveis como métodos dos objetos array.

```python
rng = np.random.default_rng(3)
y = rng.standard_normal(10)
y
```

```python
np.mean(y)
```

```python
y.mean()
```

```python
np.var(y) 
```

```python
y.var() 
```

Por default `np.var()` calcula a variância dividindo pelo número de elementos $n$ e não $n-1$.


```python
np.sqrt(np.var(y))
```

```python
np.std(y)
```

As funções `np.mean()`,  `np.var()`, e `np.std()` podem ser aplicadas apenas às linhas ou colunas de uma matriz. No exemplo a seguir construímos uma matriz $10 \times 3$ com valores amostrados de uma distribuição normal, $N(0,1)$, e calculamos a soma de suas linhas. 

```python
X = rng.standard_normal((10, 3))
X
```

Como as matrizes são ordenadas por linha, o primeiro eixo, ou seja, eixo = 0, refere-se às suas linhas. Passamos este argumento para o método `mean()` do objeto referenciado por `X` para que a operação seja realizada ao longo das linhas.

```python
X.mean(axis=0)
```

O comando a seguir gera o mesmo resultado.

```python
X.mean(0)
```

## Visualização gráfica com `matplotlib`
Em `Python`, a prática comum é usar a biblioteca `matplotlib` para fazer gráficos e visualizar imagens.
Usaremos a função `subplots()` de `matplotlib.pyplot` para criar uma figura com os
eixos nos quais plotamos nossos dados. Para mais detalhes sobre como fazer gráficos em `Python` 
visite [matplotlib.org/stable/gallery/](https://matplotlib.org/stable/gallery/index.html).

Com `matplotlib`, um gráfico consiste em uma *figura* (*figure*) e um ou mais *eixos* (*axes*). 
Pense na figura como uma tela em branco sobre a qual um ou mais gráficos serão exibidos. Ou seja, 
a figura é a janela de plotagem. Os *eixos* contêm informações importantes sobre cada gráfico, 
como os rótulos, título e muito mais. Como veremos, na biblioteca `matplotlib` a palavra
*eixos* não é o plural de *eixo*, pois os *eixos* de um gráfico contêm muito mais informações do que 
apenas o eixo $x$ e o eixo $y$.

Para plotar um gráfico, começamos importando `matplotlib.pyplot`. Desse pacote usamos a função
`subplots()` durante a criação de figuras. Essa função retorna uma tupla de comprimento dois: 
o objeto **figure**, bem como o objeto **axes**. Normalmente, passaremos `figsize` como argumento para definir o tamanho da figura.
Tendo criado nossos eixos, plotamos nosso primeiro gráfico usando o método `plot()`.
Para mais informações digite `ax.plot?`.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(8, 8))
x = rng.standard_normal(100)
y = rng.standard_normal(100)
ax.plot(x, y)
plt.show()
```

A plotagem anterior produziu um gráfico com os pontos conectados por linhas, que é o padrão, mas os pontos não estão destacados. 
O exemplo abaixo destaca os pontos.

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x, y, marker='o')
plt.show()
```

Para criar um gráfico de dispersão (scatter plot), usamos `ax.scatter()`.

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(x, y, marker='o')
plt.show()
```

Para rotular o gráfico, usamos os métodos `set_xlabel()`,  `set_ylabel()`, e  `set_title()` do objeto `axes`.

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(x, y, marker='o')
ax.set_xlabel("eixo x")
ax.set_ylabel("eixo y")
ax.set_title("X vs Y");
plt.show()
```

Como temos acesso ao objeto `figure` podemos alterar alguns aspectos e então rexibi-lo novamente. Por exemplo, a seguir mudamos o tamanho de `(8, 8)` para `(12, 3)`.

```python
fig.set_size_inches(12,3)
fig
```

Ocasionalmente, desejaremos criar vários gráficos dentro de uma única figura. Isso pode ser
feito passando argumentos adicionais para `subplots()`. Abaixo criamos uma grade de 
gráficos $2 \times 3$ em uma figura cujo tamanho é determinado pelo argumento `figsize`. Nesses arranjos, muitas vezes queremos que todos os gráficos tenham um eixo comum. O método `ax.contour()` produz um gráfico com curvas de nível 
para representar superfícies, semelhante a um
mapa topográfico.  São necessários três argumentos:

* Um vetor de valores `x` (a primeira dimensão),
* Um vetor de valores `y` (a segunda dimensão) e
* Uma matriz cujos elementos correspondem ao valor `z` (a terceiro
dimensão) para cada coordenadas `(x,y)`.

Para criar `x` e `y`, usaremos o comando `np.linspace(a, b, n)`, 
que retorna um vetor de `n` números começando em `a` e terminando em `b`.

```python
fig, ax = plt.subplots(figsize=(8, 8))
x = np.linspace(-np.pi, np.pi, 50)
y = x
f = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
ax.contour(x, y, f);
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.contour(x, y, f, levels=45);
```

A notação de **slice** (fatiamento) é usada para indexar sequências tais como listas, tuplas and arrays.
Suponha que queiramos recuperar do quarto ao sexto (inclusive) elemento de uma string. Nós obtemos um slice 
da string usando a notação de indexação `[3:6]`.

No bloco de código acima, a notação `3:6` é uma abreviação para `slice(3,6)` quando usada dentro de

```python
"hello world"[slice(3,6)]
```

Você poderia esperar que slice(3,6) produzisse do quarto ao sétimo caracteres na string de texto (lembrando que o Python começa sua indexação em zero), mas em vez disso ele produz do quarto ao sexto. Isso também explica por que o comando np.arange(0, 10) anterior gera apenas os números inteiros de $0$ a $9$.

## Indexando dados
Para começar, vamos criar um array `numpy` bidimensional.

```python
A = np.array(np.arange(16)).reshape((4, 4))
A
```

Se digitarmos `A[1,2]` recuperamos o elemento correspondente à segunda linha e terceira coluna. 

```python
A[1,2]
```

O primeiro número após o símbolo de colchete `[`
refere-se à linha e o segundo número refere-se à coluna.

### Indexação de linhas, colunas e submatrizes
 Para selecionar várias linhas de uma vez, podemos passar uma lista. Por exemplo, `[1,3]` recuperará a segunda e a quarta linhas:

```python
A[[1,3]]
```

Para selecionar a primeira e a terceira colunas, passamos [0,2] como segundo argumento entre colchetes. Neste caso precisamos fornecer o primeiro como primeiro argumento `:`,  que seleciona todas as linhas.

```python
A[:,[0,2]]
```

Agora, suponha que queiramos selecionar a submatriz composta pela segunda e quarta 
linhas, bem como a primeira e a terceira colunas. É aqui que
a indexação fica um pouco complicada. É natural tentar usar listas para recuperar as linhas e colunas:

```python
A[[1,3],[0,2]]
```

 Oops --- o que aconteceu? Obtivemos uma matriz unidimensional de comprimento dois idêntica a

```python
np.array([A[1,0],A[3,2]])
```

Da mesma forma, o código a seguir não consegue extrair a submatriz composta pela segunda e quarta linhas e pela primeira, terceira e quarta colunas:

```python
A[[1,3],[0,2]]
```

Podemos ver o que deu errado aqui. Quando são fornecidas duas listas de indexação, a interpretação numpy é que elas fornecem pares de índices para uma série de entradas. É por isso que o par de listas deve ter o mesmo comprimento. Porém, essa não era nossa intenção, pois estamos procurando uma submatriz.

Uma maneira fácil de fazer isso é a seguinte. Primeiro criamos uma submatriz subdividindo as linhas de `A` e, em seguida, criamos uma submatriz adicional subdividindo suas colunas.


```python
A[[1,3]][:,[0,2]]
```

Entretanto, existem maneiras mais eficientes de alcançar o mesmo resultado.

A função de `np.ix_()` nos permite extrair uma submatriz usando listas.

```python
idx = np.ix_([1,3],[0,2,3])
A[idx]
```

Alternativamente, podemos obter subconjuntos de matrizes de forma eficiente usando slices.

O slice 1:4:2 captura o segundo e o quarto itens de uma sequência, enquanto a fatia 0:3:2 captura o primeiro e o terceiro itens (o terceiro elemento em uma sequência de fatia é o tamanho do passo).

```python
A[1:4:2,0:3:2]
```

Por que podemos recuperar uma submatriz diretamente usando slices, mas não usando listas? É porque eles são tipos diferentes de Python e são tratados de maneira diferente por numpy. As fatias podem ser usadas para extrair objetos de sequências arbitrárias, como strings, listas e tuplas, enquanto o uso de listas para indexação é mais limitado.

### Indexação Booleana
O tipo *Boolean* possui dois valores possíveis:  `True` ou `False` (também representado como $1$ e $0$, respectivamente).
A seguir criamos um vetor de $0$'s, representados como Booleanos, de comprimento igual à primeira dimensão de `A`.

```python
keep_rows = np.zeros(A.shape[0], dtype=np.bool_)
keep_rows
```

Agora definimos dois dos elementos como `True`.

```python
keep_rows[[1,3]] = True
keep_rows
```

Observe que os elementos de `keep_rows`, quando vistos como números inteiros, são iguais aos
valores de `np.array([0,1,0,1])`. Abaixo, usamos o operador `==` para verificar se são realmente iguais. Quando aplicada a dois arrays, a operação `==` é aplicada elemento a elemento.

```python
np.all(keep_rows == np.array([0,1,0,1]))
```

(Aqui, a função np.all() verificou se todas as entradas do array resultante são verdadeiras. Uma função semelhante, np.any(), pode ser usada para verificar se alguma entrada de um array é verdadeira.)

Agora observe a diferença nos comandos abaixo. Primeiro, um array de inteiros é passado para selecionar as linhas de A.vPor outro lado, keep_rows recupera apenas a segunda e a quarta linhas de A --- ou seja, as linhas para as quais o booleano é igual a TRUE.

```python
A[np.array([0,1,0,1])]
```

```python
A[keep_rows]
```

Este exemplo mostra que booleanos e inteiros são tratados de forma diferente por numpy.

Novamente usamos a função np.ix_() para criar uma malha contendo a segunda e quarta linhas e a primeira, terceira e quarta colunas. Desta vez, aplicamos a função a booleanos, em vez de listas.

```python
keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2, 3]] = True
idx_bool = np.ix_(keep_rows, keep_cols)
A[idx_bool]
```

Também podemos misturar uma lista com um array de booleanos nos argumentos para np.ix_():

```python
idx_mixed = np.ix_([1,3], keep_cols)
A[idx_mixed]
```
